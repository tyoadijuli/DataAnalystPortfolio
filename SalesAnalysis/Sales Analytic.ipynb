# Add a sales column

```python
all_data = all_data[all_data['Order Date'] != 'Order Date']
all_data['Month'].unique()
```

```python
all_data['Month'] = pd.to_numeric(all_data['Month'])
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
```

### Add City Column

```python
#use .apply() method
all_data['Column'] = all_data['Purchase Address']
```

```python
def get_city (address):
    return address.split(',')[1]
# we need to extract the state or zip code as well because in the different state may have the same name of the city
def get_state (address):
    return address.split(',')[2]
all_data['City'] = all_data['Purchase Address'].apply(lambda x : get_city(x) +' ' +get_state(x))

all_data.head()
```

```python
city_sales = all_data.groupby('City').sum()
```

```python
cities = [city for city, df in all_data.groupby('City')]
plt.bar(cities, city_sales['Sales'])
plt.xticks(cities, rotation = 'vertical', size = 8)
plt.ylabel('Sales in Million USD ($)')
plt.xlabel('City')
```

### Which product is more often sold together

```python
# predict which data to analyze so it can have result of what we wanted
# 1. look for the same order Id that have same address
# 2. look for similar buying time
```

```python
df = all_data[all_data['Order ID'].duplicated(keep= False)]
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
df.head()
```

```python
#drop the duplicated value
df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head(100)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df ['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common():
    print(key,value)
    
# From this, the company can make a bette marketing strategy to increase the sales base on mos common product that sold together.
```

### What product sod the most? and why?

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']
quantity_ordered
```

```python
#visualize in chart

products = [product for product, df in product_group]
plt.bar(products, quantity_ordered)
plt.ylabel('Number Ordered')
plt.xlabel('Product')
plt.xticks(products, rotation = 'vertical', size = 8)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color = 'g')
ax2.plot(products, prices)

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quaity Ordered', color ='g')
ax2.set_ylabel('Prices ($)', color = 'blue')
ax1.set_xticklabels(products, rotation = 'vertical', size = 8)
```
